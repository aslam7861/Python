###### **Exercise 10:**
did cars consume more gas in the eighties?

```python
(
cars
  .filter(['Year', 'Combined_MPG'])
  .assign(decade = (cars['Year']//10)*10)
  .groupby(['decade'])
           .agg(avg_combined_mpg = ('Combined_MPG', 'mean'))
 
)
```

```python
#from pandas._libs.algos import unique_deltas
#from numpy.lib.arraysetops import unique
unique_models=(
    cars
        .groupby('Make')
        .agg(total_n_model = ("Model", 'nunique'))
        .reset_index()

  )
unique_models
```

```python
unique_models.merge(car_brands, how='outer', left_on='Make', right_on='brand')
```

###### **Bonus:**
create 4 groups (bins) of cars, by Year. We want to explore how cars have evolved decade by decade.

```python
# your code here
cars["Year_range_q"] = pd.qcut(cars["Year"], 
                             q=4,
                             labels=["80s", "90s", "00s", "10s"])
```

```python
 cars["Year_range"] = pd.cut(cars["Year"], 
                      bins = [1980, 1990, 2000, 2010, 2020], 
                      labels=["80s", "90s", "00s", "10s"])
```
